The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np

df = pd.read_csv('data/nobel.csv')

# Extracting the top values from 'sex' and 'birth_country'
top_sex = df['sex'].value_counts()
top_birth_country = df['birth_country'].value_counts()

# Storing the most commonly awarded gender and birth country
top_gender = top_sex.idxmax()
top_country = top_birth_country.idxmax()

top_gender, top_country
```

```python
# Creating a flag for winners whose birth country is "United States of America"
df['is_usa_winner'] = df['birth_country'] == 'United States of America'

# Creating a decade column
df['decade'] = (df['year'] // 10) * 10

# Calculating the ratio of USA winners to total winners per decade
usa_winners_per_decade = df[df['is_usa_winner']].groupby('decade').size()
total_winners_per_decade = df.groupby('decade').size()
ratio_usa_winners = usa_winners_per_decade / total_winners_per_decade

# Finding the decade with the highest ratio of US-born Nobel Prize winners
max_decade_usa = ratio_usa_winners.idxmax()

max_decade_usa
```

```python
# Creating a flag for female winners
df['is_female_winner'] = df['sex'] == 'Female'

# Calculating the ratio of female winners to total winners per decade and category
female_winners_per_decade_category = df[df['is_female_winner']].groupby(['decade', 'category']).size()
total_winners_per_decade_category = df.groupby(['decade', 'category']).size()
ratio_female_winners = female_winners_per_decade_category / total_winners_per_decade_category

# Finding the decade and category with the highest ratio of female Nobel Prize winners
max_female_decade_category = ratio_female_winners.idxmax()

# Creating the dictionary with the decade and category pair with the most female winners
max_female_dict = {
    max_female_decade_category[0]: max_female_decade_category[1]
}

max_female_dict
```

```python
# Filtering the DataFrame for female winners
female_winners = df[df['is_female_winner']]

# Finding the earliest year and corresponding category for female winners
first_female_winner = female_winners.loc[female_winners['year'].idxmin(), ['full_name', 'category']]

# Extracting the name and category of the first female winner
first_woman_name = first_female_winner['full_name']
first_woman_category = first_female_winner['category']

first_woman_name, first_woman_category
```

```python
# Counting the number of times each winner has won
winner_counts = df['full_name'].value_counts()

# Selecting those with counts of two or more
repeat_winners = winner_counts[winner_counts >= 2]

# Saving the full names as a list called repeat_list
repeat_list = repeat_winners.index.tolist()

repeat_list
```
